```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from datetime import date
import string
import pickle
```

The file `../data/stocks.csv` contains a table with 1967 rows and 4 columns that represent the quotes (stock prices) for four corporations -- Amazon, Apple, Facebook, and Microsoft -- over 1,967 days starting from 2012-05-18 and ending at 2020-03-13.

**Load the data using NumPy**.

```python
companies = sorted(['apple', 'facebook', 'amazon', 'microsoft'])
data = np.loadtxt('../data/stocks.csv', delimiter=',')
```

```python
data
```

**Re-create the plot** with all four companies on the same axis.

```python
rows = len(companies)

fig, ax = plt.subplots(1, 1, figsize=(4, 2))
for quote, company in zip(data.T, companies):
    ax.plot(quote, label=company)
ax.set(xlabel='Time', ylabel='Quote')
ax.legend();
```

**Re-create the plot** with each company on a different axis.

```python
rows = len(companies)

fig, axes = plt.subplots(rows, 1, figsize=(4, 2*rows), sharex=True, sharey=False)
for quote, company, ax in zip(data.T, companies, axes):
    ax.plot(quote, color='k')
    ax.set_title(company)
    ax.set_axis_off()
```
