# Seeds dataset with decision trees

During the session we classified the seeds dataset with KNN. Now let's try to classify it with decision trees!

```python
import warnings
warnings.simplefilter("ignore", FutureWarning)
warnings.simplefilter("ignore", DeprecationWarning)

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, LearningCurveDisplay
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

We start by loading the data.

```python
df = pd.read_csv('../data/seeds_dataset.txt', sep='\t')
df.head()
```

```python
target_name = 'cultivar'
feature_names = list(df.columns)
feature_names.remove(target_name)


X = df[feature_names].values
y = df[target_name].values
```

First, **split the data** to train and test sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

Second, **create, fit and score** a decision tree (or random forest) classifier.

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train) 
print("Accuracy = ", classifier.score(X_test, y_test))
```

**Plot a learning curve** to find out how many samples are required to train the model.

```python
LearningCurveDisplay.from_estimator(classifier, X, y, train_sizes=np.arange(0.1, 1.1, 0.1));
```
